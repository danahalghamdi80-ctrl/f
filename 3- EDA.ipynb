This project applies Exploratory Data Analysis (EDA) techniques to the Students Performance dataset. The analysis includes data overview, summary statistics, visualization, correlation analysis, and time-based trends.

# 2- Data Loading (Option 1: Load Local CSV)

```python
# Load Dataset

df = pd.read_csv("StudentsPerformance.csv")

```

## 7- Check Duplicates

```python
df.duplicated().sum()
```

## 8-Distribution of Math Score (Histogram)

```python
plt.hist(df['math score'])
plt.title("Distribution of Math Score")
plt.xlabel("Math Score")
plt.ylabel("Frequency")
plt.show()
```

Most students scored between 60 and 80 in math.

#  9- Average Math Score by Gender (Bar Chart)


```python
df.groupby('gender')['math score'].mean().plot(kind='bar')
plt.title("Average Math Score by Gender")
plt.ylabel("Average Math Score")
plt.show()
```

Male students slightly outperform female students in math on average.

## 10- Reading vs Writing Score (Scatter Plot)

```python
plt.scatter(df['reading score'], df['writing score'])
plt.title("Reading vs Writing Score")
plt.xlabel("Reading Score")
plt.ylabel("Writing Score")
plt.show()
```

There is a strong positive relationship between reading and writing scores.

# Final Conclusion



The analysis shows that gender and test preparation course influence student performance.
Female students perform better in reading and writing, while male students perform better in math.
Students who completed the preparation course achieved higher scores in all subjects.

Overall, preparation and demographic factors play an important role in academic success.
